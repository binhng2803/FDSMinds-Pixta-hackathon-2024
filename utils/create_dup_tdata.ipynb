```python
path = r'D:\Pixta\mnt\md0\projects\sami-hackathon\private\data\images\\'
for name in df[df['gender'] == 'Male']['file_name']:
    print(name)
    src = path + name
    dst = r'D:\Pixta\duplicate_male_data\images/' + name[:-4] + '_dup.jpg'
    shutil.copy(src, dst)
```
