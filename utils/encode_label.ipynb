```python
def encode_label(file_name):
    df = pd.read_csv(file_name)
    le_dict_corrected = {column: LabelEncoder().fit(classes) for column, classes in df.items()}
    
    df_new_to_encode = df.drop(['file_name', 'height', 'width', 'bbox'], axis=1)

    # Apply the previously fitted LabelEncoders to the new dataframe
    df_new_encoded = df_new_to_encode.copy()
    for column, le in le_dict_corrected.items():
        if column in df_new_encoded:
            df_new_encoded[column] = le.transform(df_new_encoded[column])

    # Adding the non-encoded columns back to the encoded dataframe
    df_new_encoded['file_name'] = df['file_name']
    df_new_encoded['height'] = df['height']
    df_new_encoded['width'] = df['width']
    df_new_encoded['bbox'] = df['bbox']
    correct_column_order = ['file_name', 'height', 'width', 'bbox', 'age', 'race', 'masked', 'skintone', 'emotion', 'gender']

    # Reorder the columns in df_new_encoded to match the specified order
    df_correct_order = df_new_encoded[correct_column_order]

    return df_correct_order

encoded_df = encode_label("../csvfile/pmale.csv")
encoded_df.to_csv("../encode_csvfile/encoded_pmale.csv", index= False)
encoded_df
```
