```python
df = pd.read_csv('labels.csv')[['file_name', 'height', 'width', 'bbox']]
df
```

```python
file_names = set(df['file_name'])
len(file_names)
```

```python
df.info()
```

```python
def yolo_label(row):
    bbox = row['bbox'][1:-1].split(',')
    bbox = list(map(float, bbox))
    x = (bbox[0]+bbox[2]/2)/row['width']
    y = (bbox[1]+bbox[3]/2)/row['height']
    w = bbox[2]/row['width']
    h = bbox[3]/row['height']
    return f'0 {x} {y} {w} {h}'
```

```python
df['yolo_label'] = df.apply(yolo_label, axis=1)
df
```

```python
df.bbox[0],df.bbox[0][1:-1].split(',')
```

```python
def yolo_label_file(f):
    return f[:-3]+'txt'
```

```python
df['yolo_label_file'] = df['file_name'].apply(yolo_label_file)
df
```

```python
yolo_file = df[['yolo_label_file', 'yolo_label']].values.tolist()
len(yolo_file)
# yolo_file
```

```python
from itertools import groupby
yolo_file.sort(key=lambda x: x[0])
grouped = groupby(yolo_file, key=lambda x: x[0])
result_dict = {key: '\n'.join(list(val[1] for val in values)) for key, values in grouped}
len(result_dict.keys()), len(result_dict.values())
```

```python
# result_dict
```

```python
LABEL_DIR = './all_yolo_labels/'
for key, val in result_dict.items():
    with open(LABEL_DIR+key, 'a') as f:
        f.write(val)
```

```python
from itertools import groupby

# Example list of tuples
your_list = [('a1', 'b1'), ('a2', 'b2'), ('a1', 'b3')]

# Sort the list by the first element of each tuple (a)
your_list.sort(key=lambda x: x[0])

# Group the tuples by the first element (a)
grouped = groupby(your_list, key=lambda x: x[0])

# Create a dictionary where keys are unique a values and values are lists of corresponding b values
result_dict = {key: '\n'.join(list(val[1] for val in values)) for key, values in grouped}

print(result_dict)
```

```python
import shutil
import os

file_list = os.listdir('yolo_dataset/valid/labels/')

# Source directory (current directory in this example)
source_directory = 'all_yolo_labels/'

# Destination directory
destination_directory = 'valid_labels/'

# Iterate over the file list and copy each file to the destination folder
for file_name in file_list:
    source_path = os.path.join(source_directory, file_name)
    destination_path = os.path.join(destination_directory, file_name)

    try:
        shutil.copy2(source_path, destination_path)
        print(f"Successfully copied {file_name} to {destination_directory}")
    except IOError as e:
        print(f"Error copying {file_name}: {e}")
```
